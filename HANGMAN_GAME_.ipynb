```python
# Hangman Game in Python ðŸ¤ 

import random

def hangman():
    # List of words to choose from
    words = ['apple', 'banana', 'grape', 'orange', 'mango', 'peach']

    # Select a random word from the list
    chosen_word = random.choice(words)

    # Create a list of underscores to represent the word
    display = ['_'] * len(chosen_word)

    # Set the number of attempts the user can make
    attempts = 6
    guessed_letters = []

    print("Welcome to Hangman!")
    print("The word has", len(chosen_word), "letters.")

    # Game loop
    while attempts > 0:
        print("\nWord to guess: " + ' '.join(display))
        print(f"Guessed letters: {', '.join(guessed_letters)}")
        print(f"Attempts left: {attempts}")

        # User input for guess
        guess = input("Enter a letter: ").lower()

        # Validate the guess
        if guess in guessed_letters:
            print("You've already guessed that letter. Try a different one.")
        elif len(guess) != 1 or not guess.isalpha():
            print("Please enter a valid single letter.")
        else:
            guessed_letters.append(guess)

            # Check if the guess is in the word
            if guess in chosen_word:
                print(f"Good job! {guess} is in the word.")
                # Reveal the correctly guessed letter
                for i in range(len(chosen_word)):
                    if chosen_word[i] == guess:
                        display[i] = guess
            else:
                print(f"Oops! {guess} is not in the word.")
                attempts -= 1

        # Check if the word is completely guessed
        if '_' not in display:
            print(f"ðŸŽ‰ Congratulations! You guessed the word: {chosen_word}")
            break

    # If the user runs out of attempts
    if attempts == 0:
        print(f"Sorry! You've run out of attempts. The word was: {chosen_word}")

# Start the game
hangman()
```
